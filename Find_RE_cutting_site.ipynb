```python
from Bio import Restriction
from Bio.Seq import Seq
```

```python
##Iput sequence

seq = ''
```

```python
## List all common restriction enzymes that cut both plasmids exactly N times

for re in Restriction.AllEnzymes:
    if not re.search(Seq(seq)):
        if re.is_blunt() and 'New England Biolabs' in re.supplier_list():
            print(re, re.elucidate())
```
